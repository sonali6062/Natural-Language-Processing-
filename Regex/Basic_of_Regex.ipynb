What is Regular Expression and Why it is used in NLP??
Regular expressions, often abbreviated as regex, are powerful tools used in text processing and pattern matching. They are a sequence of characters that define a search pattern, allowing you to match and manipulate text based on specific rules.

Regular expressions are widely used in Natural Language Processing (NLP) for various purposes:

Text Cleaning: Regular expressions are helpful in removing unwanted characters, special symbols, or punctuation marks from text data. They allow you to identify and replace patterns in the text, such as removing URLs, email addresses, or HTML tags.

Tokenization: Tokenization is the process of breaking text into smaller meaningful units called tokens. Regular expressions can be employed to split text into words, sentences, or other desired units based on specific patterns or delimiters.

Pattern Matching: Regular expressions enable you to search for specific patterns or sequences of characters within text data. This is useful for tasks like named entity recognition, finding email addresses, detecting phone numbers, or extracting dates from text.

Text Validation: Regular expressions help validate the format or structure of text data. For instance, you can use them to verify if an input matches a certain pattern, such as checking if an email address is valid or if a phone number follows a specific format.

Information Extraction: Regular expressions play a vital role in extracting specific information from text data. For example, you can use them to find and extract specific patterns like dates, addresses, or product codes from unstructured text.

Overall, regular expressions provide a flexible and powerful approach for text pattern matching and manipulation, making them a valuable tool in NLP tasks.

2.Findall

```python
# Define a string with multiple occurrences of a word
a="Hello,how are are are you?"
b='are'
# Search for the first occurrence of 'are' - demonstrates that re.search only finds the first match
re.search(b,a)
#are is present multiple times in the sentense but it is detecting only the 1st one
```

```python
# to find all the occurence of the word we use findall function
# Find all non-overlapping occurrences of pattern 'b' in string 'a'
srch=re.findall(b,a)
srch
```

#3. Finditer

```python
# to find all the occurrences and their spans, we use finditer function
# Find all non-overlapping occurrences of pattern 'b' in string 'a' and return an iterator of match objects
srch_iter = re.finditer(b, a)

# Iterate through the match objects and print the span for each
for match in srch_iter:
    print(match.span())
```

The significance of r before a string in Python (like r'\d\d\d-\d\d\d-\d\d\d') is that it creates a raw string.

This is particularly important for regular expressions because regular expressions use the backslash \ character extensively for special sequences (like \d to match a digit).

Without the r, Python would interpret the backslashes as escape sequences within the string first. This could lead to errors or force you to write double backslashes (\\) for every literal backslash you want in your regular expression.

By using a raw string (r"..." or r'...'), you tell Python to treat all backslashes within the string literally. This ensures that the regular expression engine receives the pattern exactly as intended, making your regular expressions easier to read and write correctly.

```python
# Define a string with a telephone number including hyphens
txt='My telephone number is 485-758-575'
# Define a pattern using a raw string to match the phone number format
pattern=r'\d\d\d-\d\d\d-\d\d\d'
# Search for the pattern in the string
phone_number=re.search(pattern,txt)
```

```python
# Display the match object found by re.search
phone_number
```

The pattern was found so we can do it in sorter way

```python
# Define a more concise pattern using quantifiers {3} for three digits
pattern=r'\d{3}-\d{3}-\d{3}'
```

```python
# Search for the pattern in the text using the concise pattern
re.search(pattern,txt)
```

```python
# Get and display the matched string from the search result
re.search(pattern,txt).group()
```
