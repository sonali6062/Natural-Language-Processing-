# ASCII Code to String Conversion and Vice Versa

ASCII to string conversion

```python
# Assign the string 'The quick brown fox jumps off a cliff' to the variable text
text='The quick brown fox jumps off a cliff'
# Display the value of the variable text
text
```

```python
# Iterate through each character in the string text
for i in text:
  # Print the character and its Unicode code point
  print(i,ord(i))
```

```python
# Create a list of Unicode code points for each character in the string text
ascii_=[ord(char) for char in text]
```

2-ASCII to string conversion

```python
# Display the list of ASCII code points
ascii_
```

```python
# Convert the list of ASCII code points back to a list of characters
([chr(asc) for asc in ascii_])
```

```python
# Join the list of characters into a single string
''.join([chr(asc) for asc in ascii_])
```

```python
# Get the Unicode code point for the character 'A'
ord('A')
```

```python
# Convert the ASCII code 65 back to its character representation
chr(65)#reverting back ascii code to character
```

```python
# Reverse the string 'The quick brown fox jumps off a cliff'
'The quick brown fox jumps off a cliff'[::-1]
```

```python
# Reverse the reversed string to get the original string
'The quick brown fox jumps off a cliff'[::-1][::-1]
```
